```python
#NORMLIZE PIXELS OF MASKS from 0 to 255 to 0 and 1.

import os
import cv2

# Assume mask_dir is the directory to your mask images
mask_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/mask"

# Destination directory for the normalized masks
norm_mask_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/norm_masks"
os.makedirs(norm_mask_dir, exist_ok=True)

# List all mask files
mask_files = os.listdir(mask_dir)

for mask_file in mask_files:
    # Read each mask
    mask_path = os.path.join(mask_dir, mask_file)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    if mask is None:
        continue
    
    # Convert all non-zero pixel values of the mask to 1
    mask[mask > 0] = 1
    
    # Save the normalized mask to the new directory
    norm_mask_path = os.path.join(norm_mask_dir, mask_file)
    cv2.imwrite(norm_mask_path, mask)
```

```python
""""
import os
import cv2
import matplotlib.pyplot as plt

image_dir = "/home/safeg34/IRP_DATASET/data_corsican/images"
mask_dir = "/home/safeg34/IRP_DATASET/data_corsican/masks"

# Get sorted list of image and mask files
image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.png')])
mask_files = sorted([f for f in os.listdir(mask_dir) if f.endswith('.png')])

# Rename images and masks
for i in range(len(image_files)):
    # Define new names
    new_name = f"corsican_{i+1}.png"

    # Print old and new names for image files
    old_image_name = os.path.join(image_dir, image_files[i])
    new_image_name = os.path.join(image_dir, new_name)
    print(f'Renaming image {old_image_name} to {new_image_name}')
    
    # Try to rename image file, catch any exceptions and print an error message
    try:
        os.rename(old_image_name, new_image_name)
    except Exception as e:
        print(f'Error renaming image file {old_image_name}: {e}')

    # Print old and new names for mask files
    old_mask_name = os.path.join(mask_dir, mask_files[i])
    new_mask_name = os.path.join(mask_dir, new_name)
    print(f'Renaming mask {old_mask_name} to {new_mask_name}')
    
    # Try to rename mask file, catch any exceptions and print an error message
    try:
        os.rename(old_mask_name, new_mask_name)
    except Exception as e:
        print(f'Error renaming mask file {old_mask_name}: {e}')

# Refresh image_files and mask_files after renaming
image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.png')])
mask_files = sorted([f for f in os.listdir(mask_dir) if f.endswith('.png')])

print('Images:', image_files)
print('Masks:', mask_files)

# Plot first 12 images and masks
fig, axs = plt.subplots(12, 2, figsize=(10, 30))

for i in range(12):
    # Read image and mask
    img = cv2.imread(os.path.join(image_dir, image_files[i]))
    mask = cv2.imread(os.path.join(mask_dir, mask_files[i]), cv2.IMREAD_GRAYSCALE)

    # Convert mask to 3-channel image for plotting
    mask_rgb = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)

    # Plot image and mask
    axs[i, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axs[i, 0].set_title('Image')
    axs[i, 0].axis('off')
    axs[i, 1].imshow(mask_rgb)
    axs[i, 1].set_title('Mask')
    axs[i, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
#RESIZE IMAGES AND MASKS

import os
import cv2

image_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/image"
mask_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/norm_masks"
output_image_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/resized_image"
output_mask_dir = "/home/safeg34/IRP_DATASET/SMOKE5K/test/resized_mask"

# Create the output directories if they don't exist
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_mask_dir, exist_ok=True)

# Get sorted list of image files
image_files = sorted([f for f in os.listdir(image_dir) if f.endswith('.jpg')])

# Resize images and masks
for image_file in image_files:
    # Load image
    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)

    # Resize image
    resized_image = cv2.resize(image, (512, 512))

    # Save resized image in PNG format
    output_image_path = os.path.join(output_image_dir, image_file.replace(".jpg", ".png"))
    cv2.imwrite(output_image_path, resized_image)

    # Load corresponding mask
    mask_file = image_file.replace(".jpg", ".png")
    mask_path = os.path.join(mask_dir, mask_file)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    # Resize mask
    resized_mask = cv2.resize(mask, (512, 512))

    # Save resized mask
    output_mask_path = os.path.join(output_mask_dir, mask_file)
    cv2.imwrite(output_mask_path, resized_mask)

print("Resizing complete.")
```
